```python
import pandas as pd
```

```python
data = pd.read_csv('./Analysis-Blebbistatin/annotated_deg_nona.csv')
```

```python
data
```

```python
df = data[data['Gene Symbol'].notna()]
```

```python
df
```

```python
# df.to_csv('annotated_deg.csv', index=False, sep=',')
```

```python
# Find Upregulated and Downregulated
```

```python
df_upregulated = df[(df['P.Value'] < 0.05) & (df['logFC'] > 1)].reset_index(drop=True)
df_upregulated_nopval = df[(df['logFC'] > 0.5)].reset_index(drop=True)
```

```python
df_upregulated
```

```python
df_upregulated_nopval
```

```python
df_upregulated.to_csv('annotated_deg_pos_FC1.csv', index=False, sep=',')
```

```python
df_downregulated = df[(df['P.Value'] < 0.05) & (df['logFC'] < -0.5)].reset_index(drop=True)
df_downregulated_nopval = df[(df['logFC'] < -0.5)].reset_index(drop=True)
```

```python
df_downregulated
```

```python
df_downregulated_nopval
```

```python
df_downregulated.to_csv('annotated_deg_neg_FC0.5.csv', index=False, sep=',')
```

```python
df_downregulated_3 = df[(df['P.Value'] < 0.05) & (df['logFC'] < -0.3)].reset_index(drop=True)
```

```python
df_downregulated_3
```

```python
df_downregulated_3.to_csv('annotated_deg_neg_3', index=False, sep=',')
```

```python
df_downregulated_all = df[(df['P.Value'] < 0.05) & (df['logFC'] < 0)].reset_index(drop=True)
```

```python
df_downregulated_all.to_csv('annotated_deg_neg_all.csv', index=False, sep=',')
```
